```python
import datetime, json, os, sys, urllib2

#!conda install -y plotly

import plotly
plotly.offline.init_notebook_mode(connected=True)
```

```python
dates = json.load(urllib2.urlopen('https://epic.gsfc.nasa.gov/api/natural/all'))

all_by_date = {}
for date in dates:
    date = date['date']
    url = 'https://epic.gsfc.nasa.gov/api/natural/date/{date}'.format(**locals())
    print url
    all_by_date[date] = json.load(urllib2.urlopen(url))
```

```python
x = []
y = []

for date in sorted(all_by_date.keys()):
    for image in all_by_date[date]:
        epoch_time = (datetime.datetime.strptime(image['date'], '%Y-%m-%d %H:%M:%S') - datetime.datetime(1970, 1, 1)).total_seconds()
        x.append(epoch_time)
        y.append(image['centroid_coordinates']['lon'])

scatter = plotly.graph_objs.Scatter(x = x, y = y, mode = 'markers', marker={'size': 3})
figure = plotly.graph_objs.Figure(data=[scatter], layout=plotly.graph_objs.Layout(width = 1000, height = 1000))
plotly.offline.iplot(figure, filename='scatter')
```

```python
def download_file(url, filename):
    if os.path.exists(filename):
        sys.stdout.write('%s already downloaded\n' % filename)
    else:
        if not os.path.exists(os.path.dirname(filename)):
            os.makedirs(os.path.dirname(filename))
        sys.stdout.write('Downloading %s to %s\n' % (url, filename))
        conn = urllib2.urlopen(url)
        data = conn.read()
        # Confused by apparently partial reads from www2.census.gov 7/2017.  Will this ever trigger?  Should we
        # be reading until empty and then concatenating all?
        assert(not conn.read())
        open(filename + '.tmp', "wb").write(data)
        os.rename(filename + '.tmp', filename)
        sys.stdout.write('Done, wrote %d bytes to %s\n' % (len(data), filename))
```

```python
goal_lon = -75
for date in sorted(all_by_date.keys()):
    best = None
    best_score = 4
    for image in all_by_date[date]:
        score = abs(image['centroid_coordinates']['lon'] - goal_lon)
        if score < best_score:
            best = image
    if best:
        (y,m,d) = date.split('-')
        name = best['image']
        src = 'https://epic.gsfc.nasa.gov/archive/natural/{y}/{m}/{d}/png/{name}.png'.format(**locals())
        dest = '75w/{name}.png'.format(**locals())
        download_file(src, dest)
```
